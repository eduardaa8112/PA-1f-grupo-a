```python
# escrever um programa que o usuario digita uma palavra e o programa verifica se a palavra é palindromas e imprime a palavra.

palavra = str (input("digite uma palavra: "))
invertida = ""

for i in (palavra):

  invertida = i + invertida

if invertida == palavra:
 print (f" {palavra} é palindroma")

else:
 print (f"{palavra} não é uma palavra palindroma")
```
