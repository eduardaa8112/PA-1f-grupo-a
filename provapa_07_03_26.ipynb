```python
# Estrutura de decisões

# Escrever um programa que recebe a idade do usuário e verifica se o usuário é criança, adolescente, adulto ou idoso.

# >= 12 anos é menor de idade
# =< 18 anos é adolescente
# >= Adulto

# Classificação de idade

idade = float(input("digite sua idade: "))

if idade <= 17:
  print("você é menor de idade.")

elif idade >= 18 and idade <= 59:
  print("você é adulto.")

else:
  print("você é idoso.")
```

```python
  # Desconto sobre valores
valor = float(input("Digite o valor da compra: R$"))

if valor <= 100:
    print("Sem desconto.")

elif valor <= 500:
    desconto = valor * 0.10
    valor_final = valor - desconto
    print(f"Você possui 10% de desconto. Valor final: R${valor_final:.2f}")

else:
    desconto = valor * 0.20
    valor_final = valor - desconto
    print(f"Você possui 20% de desconto. Valor final: R${valor_final:.2f}")
```

```python
# Lados de um triângulo
lado1 = float(input("Digite o primeiro lado: "))
lado2 = float(input("Digite o segundo lado: "))
lado3 = float(input("Digite o terceiro lado: "))

# Processamento
if (lado1 + lado2 > lado3) and (lado2 + lado3 > lado1) and (lado3 + lado1 > lado2):
    print("É um triangulo.")
else:
    print("Não é um triangulo")
```

```python
# Sistema de energia

energia = float(input("digite a energia do personagem: "))

if energia <= 0:
  print("Morto")

elif energia <= 30:
  print("Critico")

elif energia <= 70:
  print("Normal")

else:
  print("Saudavel")
```

```python
# media escolar

media = float(input("Digite a média final: "))

if media >= 7:
    print("Aprovado")
elif media >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```
